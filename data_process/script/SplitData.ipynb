```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('../datas/test.csv')
df = df.drop(['pos_z','vel_z'],axis=1)
df.head(20)
df['time'] = df['time'] - 1661865000
df
```

```python
FIRST=True
sample=[]
sample_cnt=0
target=[]
origin_file_path = '../datas/test.csv'
save_folder = '../datas/split_data/'



for i in range(len(df)):
  lst = df.iloc[i]
  new_time = lst[0]
  ped_id = int(lst[1])

  if FIRST:
    time = new_time
    target_id = ped_id
    target.append(lst)
    FIRST=False
    continue
  else:
    pass

  # next sample  
  if time != new_time:
    sample_cnt += 1
    time = new_time
    
    #reset
    while sample_cnt==15:
      others_data = sample
      target_data = target
      sample = []
      target = []
      sample_cnt = 0
      df_others = pd.DataFrame(others_data)
      df_others.columns = ['time', 'ID', 'pos_x', 'pos_y', 'vel_x', 'vel_y']
      df_target = pd.DataFrame(target_data)
      df_target.columns = ['time', 'ID', 'pos_x', 'pos_y', 'vel_x', 'vel_y']
      print("============")
      print(df_target)
      with open(save_folder + str(i) + '_target.csv', mode= 'a') as f:
        df_target.to_csv(f, header = True, index = False)
      FIRST = True
      
    if ped_id == target_id:
      target.append(lst)
    else:
      sample.append(lst)
  # same time
  else:
    if ped_id == target_id:
      target.append(lst)
    else:
      sample.append(lst)
    

  
```
